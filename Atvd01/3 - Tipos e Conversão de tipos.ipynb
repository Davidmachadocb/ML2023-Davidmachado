# 3 - Tipos e Conversão de tipos
O Numpy suporta uma variedade de tipos muito maior do que o Python.

Existem 5 tipos básicos numéricos
- Booleans (bool)
- Integers (int)
- Unsigned Intergers (uint)
- Floating Points (float)
- Complex

Complementarmente, a célula abaixo mostra um quadro comparativos com uma maior quantidade de tipos do Numpy.

```python
import numpy as np
```

Cria uma variavel x e nessa variavel ele passa o retorno de np.float32, que é um valor float32.

```python
x = np.float32(1.0)
x
```

Cria um array e define ele como do tipo inteiro

```python
y = np.int_([1,2,4])
y
```

Cria um array de tamanho 3 de 0 a 2 com o tipo unsigned int de 8 bits

```python
z = np.arange(3, dtype=np.uint8)
z
```

Cria uma array [1 2 3] do tipo float

```python
np.array([1, 2, 3], dtype='f')
```

## Convertendo tipos de arrays

Para convertermos arrays Numpy, basta utilizarmos o método astype().

```python
z.astype(float)
```

Pode converter também utilizando np.type, como mostrado

```python
np.int8(z)
```

Para verificarmos o tipo de um array Numpy, basta verificarmos o seu atributo dtype.

```python
z.dtype
```
