O `NumPy` é uma poderosa biblioteca Python que é usada principalmente para realizar cálculos em **Arrays Multidimensionais**.

O Código abaixo cria uma lista e depois converte para um array com o np.array. A diferença de listas e arrays é que o primeiro aceita mais de um tipo e o segundo só com um tipo.

```python
# Criação de arrays com o numpy
lista1 = [1, 2, 3]
array1 = np.array(lista1)

print(lista1)
print(array1)
print(array1.shape)
print(type(lista1))
print(type(array1))
```

Criando uma lista duas listas dentro, e depois converte para array o junto com as listas que estavam dentro. Também mostra no segundo print como podemos acessar somente uma linha do array.

```python
lista2 = [[1, 2, 3], 
          [3, 4, 5]]
array2 = np.array(lista2)

print(array2)
print('Linha 1: {}'.format(array2[0]))

print('Média de uma linha: {}'.format(np.average(array2)))
```

Realiza a operação de soma entre dois arrays, mas somando somente o array1 com o primeiro elemento (array) do array2

```python
print(array1 + array2)
```

Realiza o produto entre os arrays, nessa operação multiplicou os elementos do arrays [a b c]*[d e f] = [ad be cf] criando um novo array. Nesse caso com array1 e array2 faz primeiro array1*array2[linha 1] e depois o memsmo a linha 2, criando o novo array.

```python
print(array1 * array2)
```

Aqui é utilizado o np.memmap para criar um novo arquivo para guardar uma variavel que talvez seja muito grande pra memoria RAM durante a execução do programa, nesse caso sendo uma matriz de tamanho 1000 por 100.

```python
# Usar memory mapping quando o conteúdo não couber em memória RAM
x_mm = np.memmap('/tmp/x_mm.npy', dtype='float32', mode='w+', shape=(1000, 100))
x_mm[999][99] = 12.5
print(x_mm[999][99])
```

```python
# Delete the memmap instance to close the file (the mmap file is not deleted). 
# Deletion flushes memory changes to disk before removing the object.
del(x_mm)
```

```python
# Load
x_mm = np.memmap('/tmp/x_mm.npy', dtype='float32', mode='r', shape=(1000, 100))
print(x_mm[999][99])
```

```python
import matplotlib.pyplot as plt
```

## Pandas
Fornece estruturas de dados e funcionalidades para manipular e analisar dados de forma eficiente.

Principais estruturas de dados:
- Series - array unidimensional cujas linhas e colunas podem receber rótulos (labels).
- DataFrame - array multidimensional cujas linhas e colunas podem receber rótulos (labels).

```python
%matplotlib inline
import numpy as np
import pandas as pd
```

### Series
a função pd.series retorna um array com os valores marcados por labels, então cada elemento do array tem um label definido pra ele

```python
s1 = pd.Series([2, 7, 5, 1, 4])
print(s1)
```

Aqui é definido qual label vamos querer para o array, criando assim uma lista de rotulos e passando como parametro para pd.series no argumento index

```python
lista = [2, 7, 5, 1, 4]
rotulos = ['a', 'b', 'c', 'd', 'e']
s2 = pd.Series(lista, index=rotulos)
print(s2)
```

Podemos acessar o elemento de uma serie como um hashtable ou array normal, utilizando a posição ou label que está associado

```python
print(s1[2])
print(s2['c'])
```

Podemos plotar como grafico por meio do .plot

```python
s1.plot(marker='o', grid=True);
```

### DataFrames

Primeiro cria uma matriz np.array e em seguida cria labels para as colunas e linhas da matriz. em seguida cria um dataframe com a matriz e as labels
