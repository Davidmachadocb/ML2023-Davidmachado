```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

filename = "Latrocinio_2018_2022.xlsx"
lac_dict = pd.read_excel(filename, sheet_name=None)
```

```python
#Filtragem pegando somente a cidade de São Paulo

sp_lac = {}
year = 2018

for i in range(5):
  filter_result = lac_dict.get(str(year))['MUNICIPIO_CIRCUNSCRICAO'] == 'São Paulo'
  sp_lac[str(year)] = lac_dict.get(str(year))[filter_result]
  year = year + 1

sp_lac['2021'].head(100)
```

```python
#Convertendo as labels das colunas de maiúsculas para minúsculas

year = 2018
for i in range(5):
  sp_lac[str(year)] = sp_lac[str(year)].rename(str.lower, axis='columns')
  year = year + 1

sp_lac['2021'].head(100)
```

```python
#Os dados nas colunas estavam errados, trocados entre mês e ano, somente durante os anos de 2018 até 2021
#Corrigindo esse erro

year = 2018
for i in range(4):
  aux = sp_lac[str(year)]['mês estatistica']
  sp_lac[str(year)]['mês estatistica'] = sp_lac.get(str(year))['ano estatistica']
  sp_lac[str(year)]['ano estatistica'] = aux 
  year = year + 1
```

Plots dos anos 2018 - 2022

```python
numerical = ['idade_pessoa', 'mês estatistica', 'sexo_pessoa', 'cor_pele', 'profissao']
year = '2018'

for col in numerical:
    fig, ax = plt.subplots(figsize=(20,8))
    sns.histplot(data=sp_lac[year], x=col)
```
