```python
import pandas as pd
import folium
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering
from sklearn.metrics.pairwise import haversine_distances
from math import radians
```

```python
data = pd.read_csv('crimes.csv')
# Selecionar os atributos de latitude e longitude
X = data[['LATITUDE', 'LONGITUDE']]
X
```

```python
dbscan = DBSCAN(metric='haversine')
clusters_dbscan = dbscan.fit_predict(X)
map_dbscan = folium.Map(location=[data['LATITUDE'].mean(), data['LONGITUDE'].mean()], zoom_start=12)
for lat, lon, cluster in zip(data['LATITUDE'], data['LONGITUDE'], clusters_dbscan):
    folium.CircleMarker([lat, lon], radius=5, color='blue', fill=True, fill_color='blue').add_to(map_dbscan)
folium.LayerControl().add_to(map_dbscan)
map_dbscan.save('dbscan_clusters.html')
display(map_dbscan)
```

```python
kmeans = KMeans(n_clusters=len(set(clusters_dbscan)))
clusters_kmeans = kmeans.fit_predict(X)
map_kmeans = folium.Map(location=[data['LATITUDE'].mean(), data['LONGITUDE'].mean()], zoom_start=12)
for lat, lon, cluster in zip(data['LATITUDE'], data['LONGITUDE'], clusters_kmeans):
    folium.CircleMarker([lat, lon], radius=5, color='red', fill=True, fill_color='red').add_to(map_kmeans)
folium.LayerControl().add_to(map_kmeans)
map_kmeans.save('kmeans_clusters.html')
display(map_kmeans)
```

```python
hierarchical = AgglomerativeClustering(n_clusters=len(set(clusters_dbscan)))
clusters_hierarchical = hierarchical.fit_predict(X)
map_hierarchical = folium.Map(location=[data['LATITUDE'].mean(), data['LONGITUDE'].mean()], zoom_start=12)
for lat, lon, cluster in zip(data['LATITUDE'], data['LONGITUDE'], clusters_hierarchical):
    folium.CircleMarker([lat, lon], radius=5, color='green', fill=True, fill_color='green').add_to(map_hierarchical)
folium.LayerControl().add_to(map_hierarchical)
map_hierarchical.save('hierarchical_clusters.html')
display(map_hierarchical)
```
